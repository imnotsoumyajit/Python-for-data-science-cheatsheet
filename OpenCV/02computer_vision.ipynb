## Import Dependencies

```python
# import open cv for computer vision
import cv2
# to visualize an image in jupyter
from matplotlib import pyplot as plt
import os
```

## Connecting to webcam

```python
cap=cv2.VideoCapture(0)
# get a frame from our capture device
ret,frame=cap.read()
```

```python
ret
```

```python
plt.imshow(cv2.cvtColor(frame,cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# releases our capture
cap.release()
```

```python
def take_photo():
    cap=cv2.VideoCapture(0)
    ret,frame=cap.read()
    cv2.imwrite(os.path.join('data','output','image_capture.jpg'),frame)
    cap.release()
```

```python
take_photo()
```

## Rendering real time

```python
# connect to webcam
cap=cv2.VideoCapture(0)
# initailize a loop through every frame till we close our webcam
while cap.isOpened():
    ret,frame=cap.read()
#     show user to user
    cv2.imshow('Webcam',frame)
#     check and stop when we press q , and stop the loop
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
# release the webcam
cap.release()
cv2.destroyAllWindows()
```
