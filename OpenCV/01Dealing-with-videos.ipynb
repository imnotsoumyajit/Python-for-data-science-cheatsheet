```python
# Initialize our capture
cap=cv2.VideoCapture(os.path.join('data','videos','lion.mp4'))
# Looping through each frame
# frame_idx is frame_index (like i)
for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
#         read frame
        ret,frame=cap.read()

# gray transform
        gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
#         show image
        cv2.imshow('video Player',gray)

# waitKey(10) - to give us some time to press a key
# 0xFF - to capture what key we pressed
# Breaking out of the loop
        if cv2.waitKey(10) & 0xFF==ord('q'):
            break
    
#     Close everything
cap.release()
cv2.destroyAllWindows()
```

## Writing out videos 

```python
# Initialize our capture
cap=cv2.VideoCapture(os.path.join('data','videos','lion.mp4'))
# setup video writer
# get properties
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
fps=cap.get(cv2.CAP_PROP_FPS)
# video writer
video_writer=cv2.VideoWriter(os.path.join('data','output','outputVid.avi'),cv2.VideoWriter_fourcc('P','I','M','1'),fps,(width,height),isColor=False)


# Looping through each frame
# frame_idx is frame_index (like i)
for frame_idx in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
#         read frame
        ret,frame=cap.read()

# gray transform
        gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
#         show image
        cv2.imshow('video Player',gray)
#         write out the frame
        video_writer.write(gray)

# waitKey(10) - to give us some time to press a key
# 0xFF - to capture what key we pressed
# Breaking out of the loop
        if cv2.waitKey(10) & 0xFF==ord('q'):
            break
            
    
#     Close everything
cap.release()
cv2.destroyAllWindows()
# release video writer

video_writer.release()
```
