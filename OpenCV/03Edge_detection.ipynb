## import dependencies

```python
import cv2
import os
# for visualization  
from matplotlib import pyplot as plt
```

## load images

```python
# initialize our file path to our image
img_path=os.path.join('data','images','car.jpg')
```

```python
# read the image
img=cv2.imread(img_path)
```

```python
# with matplotlib - Easy method to visualize
plt.imshow(img)
```

 With opencv visualization - kinda hassle though

```python
# our original one ka dimensions
img.shape
```

```python
# our new width and height
(int(img.shape[1]/2),int(img.shape[0]/2))
```

```python
# resizing our image
resized_img=cv2.resize(img,(int(img.shape[1]/2),int(img.shape[0]/2)))

# without resizing it might take up all screen space 
cv2.imshow('Our Image',resized_img)
# with any key we can close it now
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## edge detection

```python
# we first need to blur as the image gets too much noise otherwise(Documentation) and make it gray , it make it clean af !
gray=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
blur=cv2.GaussianBlur(gray,(5,5),0)
```

```python
# apply the canny edge detection model
# more threshold more edges , vice versa
canny=cv2.Canny(blur,threshold1=180,threshold2=200)
```

```python
# resizing our image
resized_img=cv2.resize(canny,(int(img.shape[1]/2),int(img.shape[0]/2)))

# without resizing it might take up all screen space 
cv2.imshow('Our Image',resized_img)
# with any key we can close it now
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## export edges

```python
cv2.imwrite(os.path.join('data','output','edges.jpg'),canny)
```
