```python
# we can use 
# cv2.imshow
# but it is jankey and buggy in jupyter
# we we use matplotlib

import matplotlib 
from matplotlib import pyplot as plt
```

```python
# show img
plt.imshow(img)
# colors are diff
# as matplotlib reads in rgb while opencv reads in bgr
```

```python
# to fix this
recolor=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(recolor)
# to hide the <matplotlib.image.AxesImage at 0x20086c5cd00> thing
plt.show()
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

Write out imgs

```python
cv2.imwrite('grayimg.png',gray)
```
